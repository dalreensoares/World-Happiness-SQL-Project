Step 1 - Importing libraries and CSV files

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read the CSV files into DataFrames

```python
# df21 is the World Happiness Report 2021
df21= pd.read_csv("../Raw Data/world_happiness_report_2021.csv")
display(df21)
```

```python
# df23 is the World Happiness Report 2023
df23= pd.read_csv("../Raw Data/world_happiness_report_2023.csv")
display(df23)
```

Step 2 - Data Wrangling 

```python
# Add the column 'year' to both dataframes
df21['Year'] = 2021
df23['Year'] = 2023
```

# Drop the column 'regional indicator' from the 2021 report

```python
df21.drop(columns=['Regional indicator'], inplace=True)
```

Step 4 - Merge the two data frames

```python
merged_df = pd.concat([df21,df23])
display(merged_df)
```

Step 4 - Data Cleaning

# Create a list of the required variables 'list_of_variables'

```python
list_of_columns= ["Country name", "Ladder score", "Logged GDP per capita", "Social support", "Freedom to make life choices", "Year"]
```

# Create a new dataframe 'filtered_columns_df'

```python
filtered_columns_df = merged_df[list_of_columns]
display(filtered_columns_df.head(5).tail(5))
```

# List of countries into unique countries

```python
unique_countries = filtered_columns_df['Country name'].unique()
```

# Generate and ID for the countries by creating a dictionary comprehension

```python
country_ids = {country: idx for idx, country in enumerate(unique_countries, start=1)}
```

# Create a dataframe for country IDs

```python
country_df = pd.DataFrame(list(country_ids.items()), columns=['Country', 'ID'])
display(country_df)
```

# Map country names to IDs

```python
filtered_columns_df['Country_ID'] = filtered_columns_df['Country name'].map(country_ids) 
filtered_columns_df.drop('Country name', axis=1, inplace=True)
```

```python
display(filtered_columns_df)
```

# Reorder the columns

```python
filtered_columns_df = filtered_columns_df[["Country_ID", "Ladder score", "Logged GDP per capita", "Social support", "Freedom to make life choices", "Year"]]
```

```python
display(filtered_columns_df)
```

  # Calculate correlations for 2021 and 2023

```python
freedom_to_choose_correlation_21_query = """
WITH stats AS (
    SELECT 
        COUNT(*) AS n,
        SUM(freedom_to_choose) AS sum_x,
        SUM(ladder_score) AS sum_y,
        SUM(freedom_to_choose * ladder_score) AS sum_xy,
        SUM(freedom_to_choose * freedom_to_choose) AS sum_x2,
        SUM(ladder_score * ladder_score) AS sum_y2
    FROM 
        happiness_indicators
    WHERE 
        year = 2021
)
SELECT 
    (n * sum_xy - sum_x * sum_y) / 
    SQRT((n * sum_x2 - sum_x * sum_x) * (n * sum_y2 - sum_y * sum_y)) AS correlation
FROM 
    stats;

"""
```

```python
freedom_to_choose_correlation_21_query = pd.read_sql(freedom_to_choose_correlation_21_query, con=engine)
```

```python
display(freedom_to_choose_correlation_21_query)
```

```python
freedom_to_choose_top_bottom_10_21_query = """
(SELECT 
    'Top 10' AS category, 
    c.name AS country_name, 
    hi.ladder_score, 
    hi.freedom_to_choose
FROM 
    happiness_indicators hi
JOIN 
    countries c ON hi.country_id = c.id
WHERE 
    hi.year = 2021
ORDER BY 
    hi.ladder_score DESC
LIMIT 10)
UNION ALL
(SELECT 
    'Bottom 10' AS category, 
    c.name AS country_name, 
    hi.ladder_score, 
    hi.freedom_to_choose
FROM 
    happiness_indicators hi
JOIN 
    countries c ON hi.country_id = c.id
WHERE 
    hi.year = 2021
ORDER BY 
    hi.ladder_score ASC
LIMIT 10);

"""
```
